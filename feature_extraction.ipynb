```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
```

```python
x, sr = librosa.load('90bpm_wrld_smb_drm_8br_id_001_0949.wav', sr=44_100)
```

```python
x2, sr2 = librosa.load('100bpm_wrld_smb_drm_8br_id_001_0117.wav', sr=44_100)
```

We extract the following features from the audio files:

Temporal: Zero Crossing Rate, RMS Energy.

Spectral: Spectral Centroid, Bandwidth, Roll-Off.

MFCCs: First 13 coefficients + Delta MFCCs.

Rhythmic: Onset strength, tempo.

Harmonic: Harmonic-to-Noise Ratio.

# Zero Crossing Rate

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

```python
zero_crossings = librosa.zero_crossings( x,pad=False)
```

```python
zero_crossings.shape
```

```python
plt.figure(figsize=(14, 5))
plt.plot(x[:100])
plt.ylim(-0.0001, 0.0001)
```

```python
zcr_s = librosa.feature.zero_crossing_rate(x + 0.0001, )
plt.figure(figsize=(14, 5))
plt.plot(zcr_s[0])
```

```python
zcr_s.shape
```

```python
zcr_s_2 = librosa.feature.zero_crossing_rate(x2 + 0.0001, )
print(zcr_s_2.shape)
plt.figure(figsize=(14, 5))
plt.plot(zcr_s_2[0])
```

```python
zcr_s_2 = librosa.feature.zero_crossing_rate(x2 + 0.0001,hop_length=2048)
print(zcr_s_2.shape)
print(max(zcr_s_2[0]))

plt.figure(figsize=(14, 5))
plt.plot(zcr_s_2[0])
```
